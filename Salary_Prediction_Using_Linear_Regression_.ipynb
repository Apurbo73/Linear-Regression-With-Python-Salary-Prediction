```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df= pd.read_csv('/content/Salary Data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
x= df.drop('Salary', axis=1)
x.head()
```

```python
y= df['Salary']
y.head()
```

```python
plt.scatter(df['Experience Years'],df['Salary'],color='green')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title ("Salary Prediction Table")
```

```python
# IMporting Linear Regression:
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
```

```python
x.mean()
x
```

```python
y.mean()
```

```python

plt.scatter(x.mean(),y.mean(),color='red')
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel ('Salary')
plt.title ('Salary Prediction Table')
```

```python
reg.fit(x,y) # Training the model
m= reg.coef_
m
```

```python
c=reg.intercept_
c
```

```python
reg.predict([[16700]])
```

```python

df['Predicted Y'] = reg.predict(x)
df.head()
```

```python

plt.plot(x,reg.predict(x),color='orange')
plt.scatter(x.mean(),y.mean(),color='red')
plt.scatter(df['Experience Years'], df['Salary'])
plt.xlabel('Experience Years')
plt.ylabel ('Salary')
plt.title ('Salary Prediction Table')
```
